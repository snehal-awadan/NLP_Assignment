# Q.1) Download the data from the link below using web scrapping:


```python
import requests
from bs4 import BeautifulSoup
from urllib.request import urlopen
```

```python
url = "https://en.wikipedia.org/wiki/Sri_Lanka"
urlOpenRet = urlopen(url)
data = urlOpenRet.read().decode()
```

```python
soup = BeautifulSoup(data, 'html.parser')
text = soup.find(id = 'bodyContent').get_text(separator = '\n')
text
```

#### 1. Text Rank Summerizer

```python
from sumy.parsers.plaintext import PlaintextParser
from sumy.nlp.tokenizers import Tokenizer
from sumy.summarizers.text_rank import TextRankSummarizer 
```

```python
# initialoze parser
my_parser = PlaintextParser.from_string(text, Tokenizer('english'))

#create object
text_rank_summarizer = TextRankSummarizer()

#Extract top 3 sentences
#top 2 ranking sentences
summary = text_rank_summarizer(my_parser.document, sentences_count=3) 

for sent in summary:
    print(sent, '\n')
```

#### 2. Lex Rank Summarizer

```python
from sumy.summarizers.lex_rank import LexRankSummarizer

lex_rank_summarizer = LexRankSummarizer()

#top 3 ranking sentences:
summary = lex_rank_summarizer(my_parser.document, sentences_count=3) 

for sent in summary:
    print(sent, '\n')
```

##### 3. LSA Summarizer

```python
from sumy.summarizers.lsa import LsaSummarizer

lsa_summarizer = LsaSummarizer()

summary = lsa_summarizer(my_parser.document, sentences_count=3)

for sent in summary:
    print(sent, '\n')
```

####  Q3) Use word sense disamguiation algorithm to find the appropriate meaning of the highlighted word

```python
from nltk.wsd import lesk
from nltk.tokenize import word_tokenize
```

```python
a1=lesk(word_tokenize('He used the key to unlock the door.'),'key')
print(a1.definition())

a2=lesk(word_tokenize('The answer to the problem lies in the key details.'),'key')
print('Sentence 2 meaning-->',a2.definition())
```

```python
a3=lesk(word_tokenize('The weight is measured on a scale weighing instrument.'),'scale')
print('Sentence 1 meaning-->',a3.definition())

a4=lesk(word_tokenize('The project is too large in scale for one person.'),'scale')
print('Sentence 2 meaning-->',a4.definition())
```

```python
a5=lesk(word_tokenize('The knife has a very sharp blade.'),'sharp')
print('Sentence 1 meaning-->',a5.definition())

a6=lesk(word_tokenize('His mind was always sharp and quick.'),'sharp')
print('Sentence 2 meaning-->',a6.definition())
```

### Q4 find the bag of words using count vectorizer.

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
sentences = [
    "The postman delivered the package to the wrong address.",
    "I wrapped a beautiful present for my friend's birthday.",
    "The delivery truck arrived late due to heavy traffic.",
    "We need to check the shipping address before sending the order.",
    "Online shopping offers a wide variety of products with fast delivery."
]
```

```python
# create obj:
cvt = CountVectorizer()

new_data = cvt.fit_transform(sentences)

# convert into array:
values = new_data.toarray()

features = cvt.get_feature_names_out()
print("Bag of Words Array:",values)
```

```python
print("\nFeature Names:",features)
```
