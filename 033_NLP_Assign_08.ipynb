### Prepare and vectorize the data

```python
import re
import string

def cleanpaper(paperText):
    paperText = re.sub('http\S+\s*', ' ', paperText) # remove url
    paperText = re.sub('RT|cc', ' ', paperText) # remove RT and cc
    paperText = re.sub('#\S+', ' ', paperText) # Remove hastag
    paperText = re.sub('@\S+', ' ', paperText) # remove mention
    paperText = re.sub('[%s]'%re.escape(string.punctuation), ' ', paperText) # REMOE THE PUNCUATION
    paperText = re.sub(r'[^\x00-\x7f]',r' ', paperText)
    paperText = re.sub('\s+', ' ', paperText)

    return paperText
```

```python
cleaned = X.apply(cleanpaper)
```

```python
import string
import re
from os import listdir
from nltk.corpus import stopwords
import pandas as pd
import matplotlib.pyplot as plt
import nltk
nltk.download('stopwords')
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import  stopwords
from nltk.stem import WordNetLemmatizer
from nltk import pos_tag
```

```python
# load doc into memory
def load_doc(filename):
    # open the file as read only
    file = open(filename, 'r')
    # read all text
    text = file.read()
    # close the file
    file.close()
    return text
```

```python
data = load_doc('sarcasm-clusters.txt')
data = sent_tokenize(data)
data = pd.DataFrame(data)
data
```

```python
swords  =  stopwords.words('english')
wnl =  WordNetLemmatizer()

def lemmatize(word):
    tag =  pos_tag([word])
    pos  ='n'
    #print(tag)
    if tag[0][1].startswith('N'):
        pos ='n'
    elif tag[0][1].startswith('V'):
        pos =  'v'
    elif tag[0][1].startswith('R'):
        pos =  'r'
    elif tag[0][1].startswith('J'):
        pos =  'a'
    return wnl.lemmatize(word,pos=pos)
```

```python
def clean_text(sent):
    token1 = word_tokenize(sent)
    # Remove only the punchuations
    token2 = [token for token in token1 if token.isalnum()]
    # #) remove the stop words
    swards = stopwords.words('english')
    token3 = [token.lower() for token in token2 if token.lower() not in swards]
    token4  =  [lemmatize(token) for token in token3]
    return token4
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tf  =  TfidfVectorizer(analyzer=clean_text)
```

```python
x_new =  tf.fit_transform(data[0])
```

### Create the Clusters

```python
from sklearn.cluster  import KMeans
```

```python
km  =  KMeans(random_state=0)
```

```python
km.fit_predict(x_new)
```

```python
set(km.labels_)
```

```python
km.inertia_ ### sum sqaure error
```

### Elbow Method

```python
sse  =[]
for k in range (1,16):
    km  =  KMeans(n_clusters=k,random_state=0,n_init='auto')
    km.fit_predict(x_new)
    sse.append(km.inertia_)
```

```python
plt.xlabel('value of K ')
plt.ylabel('SSE')
plt.grid()
plt.xticks(range(1,16))
plt.plot(range(1,16),sse,marker = 'o',color  =  'r')
```
