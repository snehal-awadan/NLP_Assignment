# 1. Apply chuncking and show the chunks of type VBD>DT>

# 2. Find all the named entities present in article

```python

named_entites=nlp(text)
named_entites.ents

for ent in named_entites.ents:
    print(ent.text,'--->',ent.label_)
```

# 3. Access all the verbs and apply the porter stemmer and lancaster stemmer

```python

w_token = word_tokenize(text)
tagged_tokens = pos_tag(w_token)
tagged_tokens
```

```python
#extract all the verbs
verbs = [word[0] for word in tagged_tokens if word[1].startswith('V') ]
verbs
```

```python
#apply porterstemmer
ps = PorterStemmer()

for v in verbs:
    print(v,'---->',ps.stem(v))
```

```python
#apply lancaster stemmer

ls = LancasterStemmer()

for v in verbs:
    print(v,'---->',ls.stem(v))
```

# 4. Access all the verbs in past tense and lemmatize them. Show only unique verbs.

```python
#extract all verbs in past tense

past_verbs=set(word[0] for word in tagged_tokens if word[1]=='VBD')
past_verbs
```

```python
# lemmatize the past tense verbs

wnl = WordNetLemmatizer()

for word in past_verbs:
    print(word, '----->',wnl.lemmatize(word, pos = 'v'))
```

# 5. Read only the adjectives from the text and find their synonyms.

```python

adj = [word[0] for word in tagged_tokens if word[1].startswith('J') ]
print(adj)
```

```python
synonyms = {}

for a in adj:
    synonyms[a] = []
    for syn in wordnet.synsets(a):
        for lemma in syn.lemmas():
            synonyms[a].append(lemma.name())

# Printing synonyms
for adj, syns in synonyms.items():
    print(f"{adj}: {set(syns)}")
```

# 6. Read all the verbs and print their antonyms.

```python

antonyms = {}

for verb in verbs:
    for syn in wordnet.synsets(verb, pos=wordnet.VERB):
        for lemma in syn.lemmas():
            for antonym in lemma.antonyms():
                antonyms[verb] = antonym.name()

# Printing antonyms
for verb, antonym in antonyms.items():
    print(f"{verb}: {antonym}")
```
